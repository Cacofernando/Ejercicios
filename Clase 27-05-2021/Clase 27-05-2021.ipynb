```python
# Importación de Librerías

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import sample
from matplotlib.pyplot import figure
```

#### Significado de las columnas:

- bwt: baby's weight in ounces at birth
- gestation: duration of pregnancy in days
- parity: parity indicator (first born = 1, later birth = 0)
- age: mother's age in years
- height: mother's height in inches
- weight: mother's weight in pounds (during pregnancy)
- smoke: indicator for whether mother smokes (1=yes, 0=no)

```python
# Importación de datos

Datos = pd.read_csv("Muestra.csv", sep = ",")
```

```python
Datos.head()
```

```python
# Muestra del peso de los Bebés
Datos["bwt"]
```

```python
Muestra_Peso = []
```

```python
for i in Datos["bwt"]:
    Muestra_Peso.append(i*0.0283495)
```

```python
Muestra_Aleatoria_bwt = sample(Muestra_Peso, 30)
```

```python
len(Muestra_Aleatoria_bwt)
```

```python
plt.hist(x=Muestra_Aleatoria_bwt, cumulative=True, density = 1, bins = 100, rwidth = 1)
plt.show()
```

```python
q25, q75 = np.percentile(Muestra_Aleatoria_bwt,[.25,.75])
bin_width = 2*(q75 - q25)*len(Muestra_Aleatoria_bwt)**(-1/3)
bins = round((max(Muestra_Aleatoria_bwt) - min(Muestra_Aleatoria_bwt))/bin_width)
print("Freedman–Diaconis number of bins:", bins)
plt.hist(x=Muestra_Aleatoria_bwt, bins = bins);
```

```python

import scipy.stats as st

plt.hist(x = Muestra_Aleatoria_bwt , density=True, bins=178, label="Data")
mn, mx = plt.xlim()
plt.xlim(mn, mx)
kde_xs = np.linspace(mn, mx, 300)
kde = st.gaussian_kde(Muestra_Aleatoria_bwt)
plt.plot(kde_xs, kde.pdf(kde_xs), label="Densidad")
plt.legend(loc="upper left")
plt.ylabel('Probabilidad')
plt.xlabel('Frecuencia Peso de Bebe [KG]')
plt.title("Histograma");
```
