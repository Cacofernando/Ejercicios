```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scipy.stats import pearsonr
%matplotlib inline
```

```python
Positividad_csv = pd.read_csv('Positividad_por_comuna.csv', header = 0, sep=',')
Riqueza = pd.read_csv('Riqueza_Comuna.csv', header = 0, sep = ',')
```

```python
Positividad_csv.head()
```

```python
Riqueza.head()
```

```python
Positividad_csv = Positividad_csv[Positividad_csv['Codigo region'] == 13]
```

```python
Positividad_csv
```

```python
Positividad = Positividad_csv.dropna(axis = 0)
```

```python
Positividad_Final = Positividad[['Codigo comuna', 'positividad']]
```

```python
Positividad_Final1 = Positividad_Final.groupby(['Codigo comuna']).mean().reset_index()
```

```python
Positividad_Final1['Riqueza'] = Riqueza['HDI']
```

```python
Positividad_Final1
```

```python
print('Media y Desviación Estándar de la columna de Positividad')
print(round(np.mean(Positividad_Final1['positividad']),2))
print(round(np.std(Positividad_Final1['positividad']),2))
```

```python
print('Media y Desviación Estándar de la columna de Riqueza')
print(round(np.mean(Positividad_Final1['Riqueza']),2))
print(round(np.std(Positividad_Final1['Riqueza']),2))
```
